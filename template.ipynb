```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
full = cv2.imread('D:\Workspace\Code\openCV\dog.jpeg')
full = cv2.cvtColor(full , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face = cv2.imread('D:\Workspace\Code\openCV\dog_face.jpeg')
face = cv2.cvtColor(face , cv2.COLOR_BGR2RGB)
```

```python
method_list = ['cv2.TM_CCOEFF_NORMED' , 'cv2.TM_CCOEFF']
```

```python
for m in method_list:
    way = eval(m)    
    full_copy = full.copy()

    res = cv2.matchTemplate(full_copy , face ,method = way)
    min ,max , min_loc , max_loc = cv2.minMaxLoc(res)

    top_left = max_loc

height , width , channels = face.shape
bottom_right = (top_left[0] +width , top_left[1]+height)

cv2.rectangle(full_copy , top_left , bottom_right , (255, 0,0) , 10)
plt.subplot(121)
plt.imshow(res)
plt.title('heat map of the matching')
plt.subplot(122)
plt.imshow(full_copy)
plt.title('Detected face'):
plt.subplot(m)
plt.show()
```
